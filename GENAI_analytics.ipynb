#feature engineering


creating new columns

```python
#count missed payments across 6 months
month_cols = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']
x['total_missed'] = (x[month_cols]== 'Missed').sum(axis=1)
x['total_late'] = (x[month_cols] == 'Late').sum(axis=1)
x['total_ontime'] = (x[month_cols] == 'Om-time').sum(axis=1)
```

payment trend(is customer getting worse)

```python
#assigning score: ontime=0, late=1. missed =2
score_map = {'Om-time': 0, 'Late': 1, 'Missed': 2}
for col in month_cols :
  x[col+'_score'] = x[col].map(score_map)
#if later months score higher than earlier = getting worse
x['getting_worse'] = (x[['Month_4_score','Month_5_score','Month_6_score']].mean(axis=1) >
    x[['Month_1_score','Month_2_score','Month_3_score']].mean(axis=1)
).astype(int)
```

simple risk flag

```python
x['high_risk'] = (
    (x['Credit_Score'] < 500) &
    (x['Missed_Payments'] >= 3 ) &
    (x['Credit_Utilization'] > 0.6)
).astype(int)
```

Customer sgementation

```python
def risk_segment(row):
    score = 0
    if row['Credit_Score'] < 500: score += 3
    elif row['Credit_Score'] < 650: score += 1
    if row['Missed_Payments'] >= 4: score += 3
    elif row['Missed_Payments'] >= 2: score += 1
    if row['Credit_Utilization'] > 0.7: score += 2
    elif row['Credit_Utilization'] > 0.5: score += 1
    if row['getting_worse'] == 1: score += 2
    if row['Delinquent_Account'] == 1: score += 3

    if score >= 8: return 'High Risk'
    elif score >= 4: return 'Medium Risk'
    else: return 'Low Risk'

x['risk_segment'] = x.apply(risk_segment, axis=1)
print(x['risk_segment'].value_counts())
```

#Framework

```python
recommendations = {
    'High Risk': 'Immediate phone call + personalised payment plan offer',
    'Medium Risk': 'Automated SMS reminder + email with payment options',
    'Low Risk': 'Monitor only — no intervention needed'
}

x['recommended_action'] = x['risk_segment'].map(recommendations)

# summary table for Head of Collections
summary = x.groupby('risk_segment').agg(
    total_customers=('Customer_ID','count'),
    avg_missed_payments=('Missed_Payments','mean'),
    avg_credit_score=('Credit_Score','mean'),
    avg_loan_balance=('Loan_Balance','mean'),
    delinquency_rate=('Delinquent_Account','mean')
).round(2)

print(summary)
```
